```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df = pd.merge(df,df_country, on='Country Code', how='left')
```

```python
final_df.head(2)
```

```python
# To check data types
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
## Pie CHart- Top 3 Countries that uses Zomato
plt.pie(country_val[:3], labels=country_names[:3], autopct='%1.2f%%')
```


Observation:Zomato maximum records or transaction are from India After that USA and then United Kingdoms

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observations

1. When Rating is between 4.5 to 4.9---> Excellent
2. When Rating are between 4.0 to 3.4--->very good
3. when Rating is between 3.5 to 3.9----> good
4. when Rating is between 3.0 to 3.4----> average
5. when Rating is between 2.5 to 2.9----> average
6. when Rating is between 2.0 to 2.4----> Poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings, hue='Rating color', palette=['white','red','orange','Yellow','Green','Green'])
```

## Observation:
    1. Not rated count is very high 
    2. Max no. of rating are between 2.5 and 3.4

```python
## Count plot
sns.countplot(x='Rating color', data=ratings, palette=['blue','red','orange','Yellow','Green','Green'])
```

```python
cuisines_top = final_df[['Cuisines']].groupby('Cuisines').size().reset_index().rename(columns={0:'Count'})
```

```python
cuisines_split = final_df['Cuisines'].str.split(', ')
cuisines_flattened = cuisines_split.explode()
```

```python
cuisine_cnt = cuisines_flattened.value_counts()
```

```python
cuisine_cnt.head(10)
```

```python
sns.barplot(x=cuisine_cnt[:10].index, y=cuisine_cnt[:10].values, hue=cuisine_cnt[:10].index, palette='viridis')
```
